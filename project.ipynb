```python
#we are replacing Nan valus of Self_Employed column on the basis of applicant income
```

```python
def impute_selfemployed(cols):
    Self_Employed=cols[0]
    ApplicantIncome=cols[1]
    
    if pd.isnull(Self_Employed):
        if ApplicantIncome>4000:
            return 'Yes'
        else:
            return 'No'
    else:
        return Self_Employed
```

```python
train['Self_Employed']=train[['Self_Employed','ApplicantIncome']].apply(impute_selfemployed,axis=1)
```

```python
#Now we are replacing Nan values of loan amount by the mean values on the basis of self employment
def impute_loanamount(cols):
    LoanAmount=cols[0]
    Self_Employed=cols[1]
    
    if pd.isnull(LoanAmount):
        if Self_Employed=='Yes':
            return 130
        else:
            return 120
    else:
        return LoanAmount
```

```python
train['LoanAmount']=train[['LoanAmount','Self_Employed']].apply(impute_loanamount,axis=1)
```

```python
def impute_credit(cols):
    Credit_History=cols[0]
    Property_Area=cols[1]
    Loan_Status=cols[2]
    if pd.isnull(Credit_History):
        if (Property_Area=='Urban')&(Loan_Status=='Y'):
            return 1.0
        else:
            return 0.0
    else:
        return Credit_History
```

```python
train['Credit_History']=train[['Credit_History','Property_Area','Loan_Status']].apply(impute_credit,axis=1)
```

```python
def impute_loanamountterm(cols):
    Loan_Amount_Term=cols[0]
    Education=cols[1]
    Self_Employed=cols[2]
    if pd.isnull(Loan_Amount_Term):
        if (Education=='Graduate')&(Self_Employed=='No'):
            return 300
        else:
            return 360
    else:
        return Loan_Amount_Term
```

```python
train['Loan_Amount_Term']=train[['Loan_Amount_Term','Education','Self_Employed']].apply(impute_loanamountterm,axis=1)
```

```python
def impute_Property_Area(cols):
    
    if cols=='Urban':
        return 0
    elif cols=='Semiurban':
        return 1
    else:
        return 2
```

```python
train['Property_Area']=train['Property_Area'].apply(impute_Property_Area)
```

```python
def impute_loan_status(cols):
    
    if cols=='N':
        return 0
    else:
        return 1
```

```python
train['Loan_Status']=train['Loan_Status'].apply(impute_loan_status)
```

```python
def impute_Self_Employed(cols):
    
    if cols=='No':
        return 0
    else:
        return 1
```

```python
train['Self_Employed']=train['Self_Employed'].apply(impute_Self_Employed)
```

```python
def impute_Education(cols):
    
    if cols=='Graduate':
        return 1
    else:
        return 0
```

```python
train['Education']=train['Education'].apply(impute_Education)
```

```python
def impute_married(cols):
    
    if cols=='Yes':
        return 1
    else:
        return 0
```

```python
train['Married']=train['Married'].apply(impute_married)
```

```python
def impute_gender(cols):
    
    if cols=='M':
        return 1
    else:
        return 0
```

```python
train['Gender']=train['Gender'].apply(impute_gender)
```

```python
def impute_Dependents(cols):
    
    if cols=='3+':
        return 3
    else:
        return cols
```

```python
train['Dependents']=train['Dependents'].apply(impute_Dependents)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train,X_test,y_train,y_test=train_test_split(train.drop('Loan_Status',axis=1),train['Loan_Status'],test_size=0.30,random_state=101)
```

```python
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
pred=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
X=train[['Married', 'Dependents', 'Education','Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Loan_Amount_Term', 'Credit_History', 'Property_Area']]
y=train['Loan_Status']
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print(classification_report(y_test,pred))
```

```python
error_rate=[]
#will take some time
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
print(error_rate)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('Error rate vs K value')
plt.xlabel('K')
plt.ylabel('error rate')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train,y_train)
predict=knn.predict(X_test)
print(classification_report(y_test,predict))
```
